```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("vgsales.csv")
df
```

```python
grouped_sales_data = df.groupby('Genre')['NA_Sales','EU_Sales','JP_Sales','Other_Sales'].sum()
print(grouped_sales_data)
```

```python
normalised_sales_data = grouped_sales_data / grouped_sales_data.sum() * 100
print(normalised_sales_data)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(normalised_sales_data, annot=True, cmap='viridis')
plt.title('Heatmap of Genre by Region')
plt.xlabel('Region')
plt.show
```

Overall, action genre is the most popular across 3 regions (North America, Europe and Others), followed by sports genre.
Role-Playing game (RPG) genre is very popular in Japan.

```python
for cmap in ["YlGnBu", "coolwarm", "viridis", "magma", "RdYlBu"]:
    plt.figure(figsize=(10,6))
    sns.heatmap(normalised_sales_data, cmap=cmap, annot=True)
    plt.title(f"Heatmap with cmap = {cmap}")
    plt.show()
```
